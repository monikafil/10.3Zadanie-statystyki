```python
#import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#tworzenie danych
np.random.seed(42)  # żeby wyniki były powtarzalne

data_100 = np.random.normal(loc=3, scale=1, size=100)
data_10000 = np.random.normal(loc=3, scale=1, size=10000)
```

```python
#statystyka opisowa
stats_100 = pd.Series(data_100).describe()
stats_10000 = pd.Series(data_10000).describe()

stats_100, stats_10000
```

```python
#porownanie statystyki w jednej tabeli
comparison = pd.DataFrame({
    "n=100": stats_100,
    "n=10000": stats_10000
})
comparison
```

```python
## Porównanie statystyk i wnioski

Dla próby o liczności **n = 100** statystyki bardziej odbiegają od wartości teoretycznych
(średnia powinna wynosić 3, odchylenie 1). Jest to naturalne, ponieważ mała próba
jest bardziej podatna na przypadkowe odchylenia.

Dla próby **n = 10 000** statystyki są znacznie bliższe wartościom teoretycznym.
Działa tutaj prawo wielkich liczb — im większa próba, tym bardziej średnia i inne
statystyki zbliżają się do wartości oczekiwanych.

Na histogramach widać, że rozkład dla n=100 jest bardziej poszarpany, a dla n=10 000
gładki i przypomina idealny rozkład normalny.
```
