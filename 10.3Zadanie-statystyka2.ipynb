```python
#import
# importy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#tworzenie danych

np.random.seed(42) #zeby wyniki by;y powtarzalne

# Dane z rozkładu normalnego N(3, 1)
data_100 = np.random.normal(loc=3, scale=1, size=100)
data_10000 = np.random.normal(loc=3, scale=1, size=10000)
```

```python
#statystyka opisowa
stats_100 = pd.Series(data_100).describe()
stats_10000 = pd.Series(data_10000).describe()

stats_100, stats_10000
```

```python
#porownanie statystyki w jednej tabeli
comparison = pd.DataFrame({
    "n = 100": stats_100,
    "n = 10 000": stats_10000
})
comparison
```

```python
#histogram
plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
plt.hist(data_100, bins=30, color="skyblue", edgecolor="black")
plt.title("Histogram dla n = 100")
plt.xlabel("Wartości")
plt.ylabel("Liczność")

plt.subplot(1,2,2)
plt.hist(data_10000, bins=30, color="salmon", edgecolor="black")
plt.title("Histogram dla n = 10 000")
plt.xlabel("Wartości")
plt.ylabel("Liczność")

plt.tight_layout()
plt.show()
```

```python
#boxplot
plt.figure(figsize=(8,5))
sns.boxplot(data=[data_100, data_10000])
plt.xticks([0,1], ["n = 100", "n = 10 000"])
plt.title("Boxplot porównujący rozkłady")
plt.ylabel("Wartości")
plt.show()
```
